```python
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
from keras.layers import BatchNormalization
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import random
from keras import regularizers
import pickle
```

## Load Original Data

```python
# Assigning class name to keys of a dictionary for the titles of the images in the plot below.
labels_to_names = {
        0: 'T-shirt/top',
        1: 'Trouser',
        2: 'Pullover',
        3: 'Dress',
        4: 'Coat',
        5: 'Sandal',
        6: 'Shirt',
        7: 'Sneaker',
        8: 'Bag',
        9: 'Ankle boot',
    }
```

```python
# input image dimensions
img_rows, img_cols = 28, 28

# the data, split between train and test sets
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

x = np.concatenate((x_train, x_test))
y = np.concatenate((y_train, y_test))
sample_size = 50
rand_indexes = np.random.randint(0, len(x), sample_size)
fig, axis = plt.subplots(4, 4, figsize=(14, 14))
for i, ax in enumerate(axis.flat):
    n=random.randint(0,40000)
    ax.imshow(x_train[n])
    ax.set(title= " Original Fashion Item " + str(labels_to_names[y_train[n]]) )
  
x = x.reshape(x.shape[0], img_rows, img_cols, 1)
```

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

datagen.fit(x)
```

```python
i = 1
x_temp = x
y_temp = y
while i<5:
  d1 = datagen.flow(x, y,batch_size=25)
  x1,y1=d1.next()
  x_temp = np.append(x_temp,x1,axis=0)
  y_temp = np.append(y_temp,y1,axis=0)
  i+=1
```

```python
#pickle.dump((x_temp,y_temp), open('pickle.p', 'wb')) # to write 'arr' to disk
aug_dataset_tuple = pickle.load(open('pickle.p', 'rb')) # to read 'arr' from disk
```

```python
x, y = aug_dataset_tuple
x, x_test, y, y_test = train_test_split(x, y, test_size=0.2, random_state=13)
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=13)
```

```python
x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols)
sample_size = 50
rand_indexes = np.random.randint(0, len(x), sample_size)
fig, axis = plt.subplots(4, 4, figsize=(14, 14))
for i, ax in enumerate(axis.flat):
    n=random.randint(0,40000)
    ax.imshow(x_train[n])
    ax.set(title= " Modified Fashion Item " + str(labels_to_names[y_train[n]]) )
```
